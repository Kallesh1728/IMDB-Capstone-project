```python
import requests
from bs4 import BeautifulSoup

# Scrape 1000 data points (20 pages with 50 results each)
total_data = 1000
results_per_page = 50

# Base URL for the IMDb search page
base_url = "https://www.imdb.com/search/title/?genres=action&sort=user_rating,desc&title_type=feature&num_votes=25000,&pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=f11158cc-b50b-4c4d-b0a2-40b32863395b&pf_rd_r=XZ8X52H1R40B7KG5SNZ9&pf_rd_s=right-6&pf_rd_t=15506&pf_rd_i=top&ref_=chttp_gnr_1"

serial_number_list = []  # List to store serial numbers

for start_index in range(1, total_data + 1, results_per_page):
    # Create the URL with the appropriate "start" parameter
    url = f"{base_url}&start={start_index}"

    # Send a GET request to the webpage to get the source code of the page
    page = requests.get(url)

    # Parse the webpage content using BeautifulSoup
    soup = BeautifulSoup(page.content, 'html.parser')

    # Find all the elements containing movie names and serial numbers
    movie_elements = soup.find_all("h3", class_="lister-item-header")

    # Extract serial numbers from the list elements
    for idx, movie_element in enumerate(movie_elements, start=start_index):
        serial_number_list.append(idx)

print(serial_number_list)
```

```python
import requests
from bs4 import BeautifulSoup
import lxml

# Scrape 1000 data points (20 pages with 50 results each)
total_data = 1000
results_per_page = 50

# Base URL for the IMDb search page
base_url = "https://www.imdb.com/search/title/?genres=action&sort=user_rating,desc&title_type=feature&num_votes=25000,&pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=f11158cc-b50b-4c4d-b0a2-40b32863395b&pf_rd_r=XZ8X52H1R40B7KG5SNZ9&pf_rd_s=right-6&pf_rd_t=15506&pf_rd_i=top&ref_=chttp_gnr_1"

directors_list = []  # List to store director names

for start_index in range(1, total_data + 1, results_per_page):
    # Create the URL with the appropriate "start" parameter
    url = f"{base_url}&start={start_index}"

    # Send a GET request to the webpage to get the source code of the page
    page = requests.get(url)

    # Parse the webpage content using BeautifulSoup and the lxml parser
    soup = BeautifulSoup(page.content, 'lxml')

    # Find all the elements containing director names using XPath
    director_elements = soup.select('div.lister-item-content > p:nth-of-type(3) > a:first-of-type')

    # Get text from director elements and store them in the directors_list
    for director_element in director_elements:
        directors_list.append(director_element.text.strip())

print(directors_list)
```
