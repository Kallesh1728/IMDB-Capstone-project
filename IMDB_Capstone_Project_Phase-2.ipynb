1)  Display all the details of movies created by directors Christopher and Matt Reeves.

```python
# Assuming you have a database connection and a 'Cursor' object named 'Cursor'

# Query to retrieve movies created by directors Christopher and Matt Reeves
query = "SELECT * FROM scores WHERE Director_Name IN ('Christopher', 'Matt Reeves')"

# Execute the query and fetch all the matching rows
Cursor.execute(query)
movies_by_directors = Cursor.fetchall()

# Display the details of movies
for movie in movies_by_directors:
    sno, movie_name, director_name, duration, year, rating, metascores = movie
    print(f"Sno: {sno}")
    print(f"Movie Name: {movie_name}")
    print(f"Director Name: {director_name}")
    print(f"Duration: {duration} minutes")
    print(f"Year: {year}")
    print(f"Rating: {rating}")
    print(f"Metascores: {metascores}")
    print("--------------------")
```

2) Display all the details of movies with a duration of 140 minutes to 190 minutes.

```python
# Assuming you have a database connection and a 'Cursor' object named 'Cursor'

# Query to retrieve movies with duration between 140 and 190 minutes
query = "SELECT * FROM scores WHERE duration BETWEEN 140 AND 190"

# Execute the query and fetch all the matching rows
Cursor.execute(query)
movies_in_range = Cursor.fetchall()

# Display the details of movies
for movie in movies_in_range:
    sno, movie_name, director_name, duration, year, rating, metascores = movie
    print(f"Sno: {sno}")
    print(f"Movie Name: {movie_name}")
    print(f"Director Name: {director_name}")
    print(f"Duration: {duration} minutes")
    print(f"Year: {year}")
    print(f"Rating: {rating}")
    print(f"Metascores: {metascores}")
    print("--------------------")
```

3) Display all details of movies with ratings above 7 in ascending order.

```python
# Assuming you have a database connection and a 'Cursor' object named 'Cursor'

# Query to retrieve movies with ratings above 7 in ascending order
query = "SELECT * FROM scores WHERE rating > 7 ORDER BY rating ASC"

# Execute the query and fetch all the matching rows
Cursor.execute(query)
high_rated_movies = Cursor.fetchall()

# Display the details of movies
for movie in high_rated_movies:
    sno, movie_name, director_name, duration, year, rating, metascores = movie
    print(f"Sno: {sno}")
    print(f"Movie Name: {movie_name}")
    print(f"Director Name: {director_name}")
    print(f"Duration: {duration} minutes")
    print(f"Year: {year}")
    print(f"Rating: {rating}")
    print(f"Metascores: {metascores}")
    print("--------------------")
```

4) Display all movie names in descending order.

```python
# Assuming you have a database connection and a 'Cursor' object named 'Cursor'

# Query to retrieve all movie names in descending order
query = "SELECT movies_name FROM scores ORDER BY movies_name DESC"

# Execute the query and fetch all the matching rows
Cursor.execute(query)
movie_names_descending = Cursor.fetchall()

# Display the movie names
for movie_name in movie_names_descending:
    print(movie_name[0])
```

5) Display movie name starts with ‘P’ and their rating is greater than 7.



```python
# Assuming you have a database connection and a 'Cursor' object named 'Cursor'

# Query to retrieve movie names that start with 'P' and have a rating greater than 7
query = "SELECT movies_name FROM scores WHERE movies_name LIKE 'P%' AND rating > 7"

# Execute the query and fetch all the matching rows
Cursor.execute(query)
matching_movies = Cursor.fetchall()

# Display the movie names
for movie in matching_movies:
    print(movie[0])
```
